```
#importing the libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as pt
```

```
#reading the dataset
data = pd.read_csv('hotel.csv')
data.info()
```

```
#inspecting the head of the dataset
data.head()
```

```
#inspecting the tail of the dataset
data.tail()
```

```
data.describe(include='all')
```

```
max_count = data.describe(include='all').iloc[0,0]
col_2_drop = list()

for kolom in data.columns :
    if data[kolom].count() < max_count/2 :
        col_2_drop.append(kolom)

data = data.drop(columns=col_2_drop)
data.head()
```

```
data['agent'] = data.agent.fillna(value=data['agent'].mean())
data['country'] = data.country.fillna(value=data['country'].mode())
data['children'] = data.children.fillna(value=data['children'].mode())
data.head()
```
